```python
# OpenCV 정보 확인 예제
import cv2

# 패키지 설치 위치 확인
print(cv2.__file__)

# 패키지 버전 확인
print(cv2.__version__)
```

```python
# 평균 추적을 위한 초기 사각형 설정
track_window = None
roi_hist = None
term_crit = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 1)  # 10 = 반복횟수, 1 = 수렴 최소값
```

```python
# 동영상 파일 열기
cap = cv2.VideoCapture('./data/slow_traffic_small.mp4')
```

```python
# 첫 프레임에서 추적할 객체 선택
ret, frame = cap.read()
x, y, w, h = cv2.selectROI('Select Object', frame, False, False)
```

```python
# 추적할 객체의 초기 히스토그램 계산
roi = frame[y:y+h, x:x+w]
#cv2.imshow('roi test', roi)
#cv2.waitKey(0)

hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

cv2.imshow('roi test', hsv_roi)
cv2.waitKey(0)

roi_hist = cv2.calcHist([hsv_roi], [0], None, [180], [0,180])

cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)

# 추적할 객체 초기 윈도우 설정
track_window = (x,y,w,h)
```

```python
print(track_window)
```

```python
while True:
    # 프레임 읽기
    ret, frame = cap.read()
    if not ret:
        break
    
    # 추적할 객체의 히스토그램 역투영 계산 (유사도를 측정)
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    dst = cv2.calcBackProject([hsv], [0], roi_hist, [0,180], 1)
    
    #평균 이동 알고리즘을 통해 객체 위치 추정
    ret, track_window = cv2.meanShift(dst, track_window, term_crit)
    
    # 시각화
    x, y, w, h = track_window
    print(x, y, w, h)
    cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)
    
    cv2.imshow('Mean shift tracking', frame)
    
    # q 키누르면 종료
    if cv2.waitKey(30) & 0xff == ord('q'):
        break
        
# 자원 해제
cap.release()
cv2.destroyAllWindows
```
