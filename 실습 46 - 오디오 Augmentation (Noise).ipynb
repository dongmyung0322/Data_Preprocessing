```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
file_path = './data/blues.00001.wav'
data, sr = sf.read(file_path)

print(data,sr)
```

```python
# 노이즈 생성
noise_amp = 0.5*np.random.uniform()*np.amax(data)  # random.uniform() = 0~1사이의 랜덤한 균등분포 반환, amax() = 최대값
noise = noise_amp*np.random.normal(size=data.shape)
```

```python
# 노이즈 추가
data_noise = data.astype('float32') + noise.astype('float32')
```

```python
# 데이터 저장
save_path = './data/sample_noise_added.wav'
sf.write(save_path, data_noise, sr)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(data)
plt.title('Original')
plt.subplot(1,2,2)
plt.plot(data_noise)
plt.title('Noised')
plt.show()
```

```python
filename = './data/sample_noise_added.wav'
y, sr = sf.read(filename)
# y=시계열 sr=샘플링 속도
print(y,sr)            

ipd.Audio(y, rate=sr)
```
