```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
class HeightWeightDataset(Dataset):
    def __init__(self, csv_path):
        self.data = []
        with open(csv_path, 'r', encoding='utf-8') as f:
            next(f)  # 첫번째 라인은 해더임으로 제외
            for line in f:
                _, height, weight = line.strip().split(',')
                height = float(height)
                weight = float(weight)
                convert2kg_data = round(self.convert2kg(weight),2)
                convert2cm_data = round(self.convert2cm(height),1)
                
                self.data.append([convert2cm_data, convert2kg_data])
            print(self.data)
        
    def __getitem__(self, index):
        data = torch.tensor(self.data[index], dtype=torch.float)
        return data
        
    def __len__(self):
        return len(self.data)
        
    def convert2kg(self, weight):
        return weight * 0.453592
    
    def convert2cm(self, height):
        return height * 2.54
```

```python
dataset = HeightWeightDataset('./data/hw_200.csv')
dataLoader = DataLoader(dataset, 1, shuffle=True)
```

```python
for batch in dataLoader:
    print(batch,batch.shape)
    x = batch[:,0].unsqueeze(1)
    y = batch[:,1].unsqueeze(0)
    print(x,y)
    print(x.shape, y.shape)
```
