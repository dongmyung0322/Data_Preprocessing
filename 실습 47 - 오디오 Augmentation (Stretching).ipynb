```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
import IPython.display as ipd
```

```python
data, sr = librosa.load('./data/blues.00001.wav')
```

```python
# stretching audio
data_stretched = librosa.effects.time_stretch(data, rate=0.8) # rate = %만큼 오디오를 늘려줌 (0.8 = 80%)
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
librosa.display.waveshow(data, sr=sr, alpha=0.8) # alpha = 투명도
plt.title('Original')

plt.subplot(1,2,2)
librosa.display.waveshow(data_stretched, sr=sr, alpha=0.8)
plt.title('Stretch')

plt.show()
```

```python
save_path = './data/sample_stretched.wav'
sf.write(save_path, data_stretched, sr)

filename = './data/sample_stretched.wav'
y, sr = sf.read(filename)
          

#ipd.Audio(data, rate=sr)
ipd.Audio(y, rate=sr)
```
