```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
def expand2square(pil_img, background_color):
    width, height = pil_img.size
    
    if width == height:
        return pil_img
    elif width > height:
        result = Image.new(pil_img.mode, (width,width), background_color) # 백그라운드 이미지 생성
        result.paste(pil_img, (0, (width-height) // 2)) # 생성된 이미지 가운대 붙이기
        return result
    else:
        result = Image.new(pil_img.mode, (height,height), background_color)
        result.paste(pil_img, ((height - width) // 2), 0)
        return result
```

```python
def resize_with_padding(img, new_size):
    img = img.resize((new_size[0], new_size[1]), Image.ANTIALIAS)
    return img
```

```python
img = Image.open('./data/cat.png')

img_after = expand2square(img,(0,0,0))

img_after2 = resize_with_padding(img_after,(150,150))

plt.imshow(img_after)
plt.show()

plt.imshow(img_after2)
plt.show()
```
