```python
#! pip install natsort
import os
from natsort import natsort
import glob
```

# 이미지 폴더의 파일 가져오기 (os.listdir)

```python
img_dir = './img_download/'

file_list = os.listdir(img_dir)
file_list = sorted(file_list)  # 앞의 문자열 기반
print(file_list)
```

```python
# 숫자에 따른 정렬
file_list2 = natsort.natsorted(os.listdir(img_dir))
print(file_list2)
```

# 이미지 폴더의 파일 가져오기 (glob.glob)

```python
file_list = glob.glob(os.path.join(img_dir, '*.png'))
file_list2 = natsort.natsorted(file_list)
print(file_list2)
```

# 이미지 폴더의 파일 가져오기 (os.walk)

```python
def get_img_path(root_path):
    file_paths = []
    for (path, dir, files) in os.walk(root_path):
        print('path: ', path)
        print('dir: ', dir)
        print('files: ', files)
        
        for file in files:
            ext = os.path.splitext(file)[-1].lower()  # .png등만 가져옴
            formats = ['.bmp','.jpg','.jpeg','.png','.tif','.tiff','.dng']
            if ext in formats:
                file_path = os.path.join(path, file)
                print(file_path)
                file_paths.append(file_path) 
        print(file_paths)
        
get_img_path(img_dir)
```
