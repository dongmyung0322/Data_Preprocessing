```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample_img = cv2.imread('./data/ex_image.png')
sample_logo = cv2.imread('./data/ex_image_logo.png')

sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)
sample_logo = cv2.cvtColor(sample_logo, cv2.COLOR_BGR2RGB)

plt.imshow(sample_img)
plt.show()

plt.imshow(sample_logo)
plt.show()
```

```python
img1 = cv2.resize(sample_img, (800,600))
img2 = cv2.resize(sample_logo, (800,600))

print(img1.shape, img2.shape)

plt.imshow(img1)
plt.show()

plt.imshow(img2)
plt.show()
```

```python
# 
blended = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

plt.imshow(blended)
plt.show()
```

```python
#   
blended = cv2.addWeighted(img1, 0.8, img2, 0.2, 0)

plt.imshow(blended)
plt.show()
```

```python
blended = cv2.addWeighted(img1, 5, img2, 5, 0)

plt.imshow(blended)
plt.show()
```

```python
img_rec = np.ones((400,400), dtype='uint8')
cv2.rectangle(img_rec, (50,50), (300,300), (255,255,255), -1)

plt.imshow(img_rec,'gray')
plt.show()
```

```python
img_circle = np.ones((400,400), dtype='uint8')
cv2.circle(img_circle, (300,300), 50, (255,255,255), -1)

plt.imshow(img_circle,'gray')
plt.show()
```

# cv2.bitwise and,or,not 

```python
bitAnd = cv2.bitwise_and(img_rec, img_circle)
plt.imshow(bitAnd,'gray')
plt.show()

bitOr = cv2.bitwise_or(img_rec, img_circle)
plt.imshow(bitOr,'gray')
plt.show()

bitXor = cv2.bitwise_xor(img_rec, img_circle)
plt.imshow(bitXor,'gray')
plt.show()

bitNot_rec = cv2.bitwise_not(img_rec)
plt.imshow(bitNot_rec,'gray')
plt.show()

bitNot_circle = cv2.bitwise_not(img_circle)
plt.imshow(bitNot_circle,'gray')
plt.show()
```
