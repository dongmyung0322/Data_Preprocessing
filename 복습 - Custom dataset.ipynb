```python
import os
import glob
import albumentations as A
import cv2
from torch.utils.data import Dataset, DataLoader
```

```python
class myCustomdataset(Dataset):
    
    def __init__(self, img_path, transform):
        self.img_paths = glob.glob(os.path.join(img_path,'*','*','*.png'))
        self.transform = transform
        self.label_dict = {'Stealing_Courier':0}
    
    def __getitem__(self, index):
        
        img_path = self.img_paths[index]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        
        label_temp = img_path.split('\\')[1]
        label = self.label_dict[label_temp]
        
        # Albumentation 사용시
        if self.transform:
            img = self.transform(img)['image']
        # pytorch 사용시
        #if self.transform:
            #img = self.transform(img)
            
        return img, label
    
    def __len__(self):
        return len(self.img_paths)
        
test = myCustomdataset('./AI_hub_assignment/',transform=None)

#  __getitem__ 값 보기위해 필요
for i in test:
    pass
```
